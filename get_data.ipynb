```python
SEASONS = list(range(1991,2026))
```

```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
dfs = []
for SEASON in SEASONS:
    with open("mvp/{}.html".format(SEASON), "r", encoding="utf-8") as f:
            page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="over_header").decompose()
    mvp_table = soup.find(id="mvp")
    mvp = pd.read_html(str(mvp_table))[0]
    mvp["Year"] = SEASON
    
    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)
```

```python
mvps.to_csv("mvps.csv")
```

```python
dfs = []
for SEASON in SEASONS:
    with open("player_stats/{}.html".format(SEASON), "r", encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="thead").decompose()
    soup.find('tr', class_="norank").decompose()
    player_table = soup.find(id="per_game_stats")
    player = pd.read_html(str(player_table))[0]
    player["Year"] = SEASON
    dfs.append(player)
```

```python
players = pd.concat(dfs)
```

```python
players.to_csv("player.csv")
```

```python
dfs = []
for SEASON in SEASONS:
        with open("team_standings/{}.html".format(1991), "r", encoding="utf-8") as f:
                page = f.read()

        soup = BeautifulSoup(page, "html.parser")
        soup.find('tr', class_="thead").decompose()
        team_table = soup.find(id="divs_standings_E")
        team = pd.read_html(str(team_table))[0]
        team["Year"] = SEASON
        team["Team"] = team["Eastern Conference"]
        del team["Eastern Conference"]
        dfs.append(team)

        soup = BeautifulSoup(page, "html.parser")
        soup.find('tr', class_="thead").decompose()
        team_table = soup.find(id="divs_standings_W")
        team = pd.read_html(str(team_table))[0]
        team["Year"] = SEASON
        team["Team"] = team["Western Conference"]
        del team["Western Conference"]
        dfs.append(team)
```

```python
teams = pd.concat(dfs)
```

```python
teams.to_csv("teams_stats.csv")
```
