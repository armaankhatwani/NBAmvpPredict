```python
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found+=1
            ps.append(found/seen)
        seen+=1
    return sum(ps)/ len(ps)
```

```python
def add_ranks(combination):
    combination = combination.sort_values("Share", ascending=False)
    combination["Rk"] = list(range(1,combination.shape[0]+1))
    combination = combination.sort_values("predictions", ascending=False)
    combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
    combination["Diff"] = combination["Rk"] - combination["Predicted_Rk"]
    return combination
```

```python
def is_mvp_eligible(row):
    if row["Year"] < 2023:
        return True  # no hard rule
    return row["G"] >= 65  # simple approximation
    
stats = stats[stats.apply(is_mvp_eligible, axis=1)]
```

```python
def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    for year in years[5:]:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors], train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share", "Year"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps)/len(aps), aps, pd.concat(all_predictions)
```

```python
import numpy as np
# add new features to df

df = stats.copy()
# --- Shooting efficiency ---
# True Shooting % = PTS / (2 * (FGA + 0.44 * FTA))
df["TS%"] = df["PTS"] / (2 * (df["FGA"] + 0.44 * df["FTA"]))
df["TS%"] = df["TS%"].replace([np.inf, -np.inf], np.nan).fillna(0)

# Rank teams by win percentage within each season
team_ranks = (
    df.groupby(["Year", "Tm"])["W/L%"]
      .first()  # one value per team-year
      .groupby("Year")
      .rank(ascending=False)
)

# Map back to player rows
df["Team_WinRank"] = df.set_index(["Year", "Tm"]).index.map(team_ranks)

# Mark top 2 seeds
df["TopSeed"] = (df["Team_WinRank"] <= 2).astype(int)
```

```python
def top1_top5_test(predict):
    top1_correct = 0
    top5_correct = 0
    for year, group in predict.groupby("Year"):
        actual_mvp_rank = group.loc[group["Rk"] == 1, "Player"].values[0]
        predicted_ranks = group.set_index("Player")["Predicted_Rk"]
        predicted_rank_of_mvp = predicted_ranks[actual_mvp_rank]

        # Top-1
        if predicted_rank_of_mvp == 1:
            top1_correct += 1
        # Top-5
        if predicted_rank_of_mvp <= 5:
            top5_correct += 1

    n_years = predict["Year"].nunique()
    return top1_correct / n_years, top5_correct / n_years
```

```python
def top10_mean_abs_rank_diff(predict):
    top_players = predict[predict["Rk"] <= 10]
    mean_top10_diff = top_players["Diff"].abs().mean()
    return mean_top10_diff
```

```python
import xgboost as xgb
import pandas as pd
import numpy as np

def backtest_xgb(stats, years, predictors, params, num_boost_round=300):
    aps = []
    all_predictions = []

    for year in years[5:]:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]

        X_train = train[predictors]
        y_train = train["Share"].values
        groups_train = train.groupby("Year").size().to_list()

        dtrain = xgb.DMatrix(X_train, label=y_train)
        dtrain.set_group(groups_train)

        model = xgb.train(params, dtrain, num_boost_round=num_boost_round)

        X_test = test[predictors]
        dtest = xgb.DMatrix(X_test)
        preds = model.predict(dtest)

        combination = test[["Player", "Share", "Year"]].copy()
        combination["predictions"] = preds

        combination = add_ranks(combination)

        aps.append(find_ap(combination))

        all_predictions.append(combination)

    mean_ap = sum(aps)/len(aps)
    return mean_ap, aps, pd.concat(all_predictions)
```

```python
# XGBoost LambdaRank parameters
params = {
    "objective": "rank:pairwise",
    "eval_metric": "ndcg@5",
    "eta": 0.1,
    "max_depth": 6,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "tree_method": "hist"
}
```

```python
# Test without df predicitors
xgb_mean_ap, xgb_aps, xgb_predictions = backtest_xgb(stats, years, predictors, params, num_boost_round=300)
xgb_mean_ap
```
